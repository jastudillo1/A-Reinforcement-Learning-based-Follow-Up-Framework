```python
from glob import glob
import json
import numpy as np
import pandas as pd
```

```python
results_dir = './results'
```

```python
def get_stats(summ_path, sett_path):
    summ = pd.read_csv(summ_path)
    with open(sett_path) as f:
        sett = json.load(f)
    
    acc = np.mean(summ.prediction==summ.label)
    n_photo = summ.n_photo.mean()
    n_spec = summ.n_spec.mean()
    n_color = summ.n_color.mean()
    prob = summ.probability.mean()
    c_g = sett['tree_args']['cost_obs']
    c_s = sett['tree_args']['cost_spec']
    c_c = sett['tree_args']['cost_col']
    k = sett['tree_args']['cost_coeff']
    
    n = summ.shape[0]
    queries_g = sum(summ['n_photo']-5)
    queries_s = sum(summ['n_spec'])
    queries_c = sum(summ['n_color'])
    queries = queries_g+queries_s+queries_c
    share_g = queries_g/queries
    share_s = queries_s/queries
    share_c = queries_c/queries
    
    return {'acc': acc, 'mean_g': n_photo, 'mean_s':n_spec, 'mean_c':n_color, 'mean_prob':prob, 
            'c_g':c_g, 'c_s':c_s, 'c_c':c_c, 'k':k, 'share_g':share_g, 'share_s':share_s, 'share_c':share_c,
            'queries_g':queries_g, 'queries_s':queries_s, 'queries_c':queries_c, 'queries': queries,
            'dataset_n':n
           }
```

```python
summ_paths = glob(f'{results_dir}/summary*.csv')

stats = []
for summ_path in summ_paths:
    i = summ_path.rstrip('.csv').split('summary')[-1]
    sett_path = f'{results_dir}/settings{i}.json'
    stats.append(get_stats(summ_path, sett_path))

stats = pd.DataFrame(stats)
```
