```python
from glob import glob
import json
import pandas as pd
```

```python
dir_ = './strategies'
```

```python
rwrd_paths = glob(dir_+'/reward*.csv')
```

```python
info = []
for rwrd_path in rwrd_paths:
    rwrd = pd.read_csv(rwrd_path)
    lt_brute = rwrd[rwrd['strategy']=='reward_greedy']['lt-brute ocurrences'].values[0]
    sett_path = rwrd_path.split('reward')
    sett_path = 'settings'.join(sett_path).rstrip('csv')
    sett_path = sett_path + 'json'
    
    info_i = {'lt-brute': lt_brute, 'reward_path': rwrd_path, 'settings_path': sett_path}
    info.append(info_i)
```

```python
info = pd.DataFrame(info)
info = info.sort_values('lt-brute', ascending=False, axis=0)
```

```python
info.head()
```

```python
counter = 0
sett_path = info.iloc[0]['settings_path']
for _, row in info.iterrows():
    if counter >=10:
        break
    lt_brute = row['lt-brute']
    sett_path = row['settings_path']
    with open(sett_path) as f:
        sett = json.load(f)
    
    print(sett_path)
    print('lt-brute: ', lt_brute)
    print(sett)
    print('------------')
    counter +=1
```
