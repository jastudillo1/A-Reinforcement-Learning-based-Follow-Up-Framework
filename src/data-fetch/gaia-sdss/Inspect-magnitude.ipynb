**Purpose**

Compare mean magnitudes of all objects in Gaia DR2 to those which are in the cross-match with SDSS-DR14. 

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ctlg_dir = '../../../data/catalogues'

gaia_all_path = f'{ctlg_dir}/gaia/gaia-variable-sources.csv'
gaia_x_path = f'{ctlg_dir}/gaia-sdss/cross-match.csv'
```

```python
gaia_all = pd.read_csv(gaia_all_path)
gaia_all = gaia_all.set_index('source_id')
gaia_x = pd.read_csv(gaia_x_path)
gaia_x = gaia_x.set_index('source_id_gaia')
gaia_x = gaia_x.join(gaia_all)
```

```python
g_mag_all = gaia_all['phot_g_mean_mag']
g_mag_x = gaia_x['phot_g_mean_mag']

plt.figure(figsize=(5,3), dpi=200)
plt.hist(g_mag_all, color='red', density=True, label='Gaia Variable Sources', alpha=0.5)
plt.hist(g_mag_x, color='blue', density=True, label='Gaia Variable Sources in SDSS X-match', alpha=0.5)

plt.xlabel('Mean G band magnitude')
plt.ylabel('Normalized counts')
plt.title('Gaia v/s Gaia in cross-match:\n Comparison of mean G band magnitude histograms')
plt.legend()
plt.show()
```
