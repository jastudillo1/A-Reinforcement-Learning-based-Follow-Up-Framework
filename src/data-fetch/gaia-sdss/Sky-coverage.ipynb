```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ctlg_dir = '../../../data/catalogues'
gaia_path = f'{ctlg_dir}/gaia/gaia-coords.csv'
sdss_path = f'{ctlg_dir}/sdss/sdss-coords.csv'
```

```python
gaia_coords_df = pd.read_csv(gaia_path)
sdss_coords_df = pd.read_csv(sdss_path)

print(gaia_coords_df.shape, sdss_coords_df.shape)
```

```python
gaia_ra = gaia_coords_df['ra'].values
gaia_dec = gaia_coords_df['dec'].values

sdss_ra = sdss_coords_df['PLUG_RA'].values[::4]
sdss_dec = sdss_coords_df['PLUG_DEC'].values[::4]
```

```python
plt.rcParams.update({'font.size': 14})
plt.figure(figsize=(10,10))
plt.scatter(sdss_ra, sdss_dec, c='blue', s=1, label='SDSS DR14', alpha=0.2)
plt.scatter(gaia_ra, gaia_dec, c='red', s=1, label='Gaia DR2', alpha=0.2)
plt.xlabel('Right Ascension')
plt.ylabel('Declination')
plt.title('Sky Coverage Datasets')
plt.legend()
plt.show()
```
